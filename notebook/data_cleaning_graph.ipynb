```python
df['duration'] = df['duration'].replace(r'^\s*$', '0', regex=True)
df['duration'].head(5)
```

```python
df['duration'] = df['duration'].fillna(0.0).astype(int)
```

```python
for i in ['duration', 'listeners', 'playcount']:
    df[i] = df[i].replace(r'^\s*$', '0', regex=True)
    df[i] = df[i].fillna(0.0).astype(int)
df.info()
df.shape
```

```python
tg = []
imgl = []
imgxl = []
for i, r in df.iterrows():
    tag = r['tag']
    image = r['image']
    t = []
    il = 'n/a'
    iexl = 'n/a'
    for j in tag:
        t.append(j['name'])
    for k in image:
        if k['size'] == 'large':
            il = k['#text']
        if k['size'] == 'extralarge':
            iexl = k['#text']
    tg.append(t)
    imgl.append(il)
    imgxl.append(iexl)
print(len(tg), len(imgl), len(imgxl))
```

```python
df['tag'] = tg
df['image_url_large'] = imgl
df['image_url_exlarge'] = imgxl
df = df.drop('image', axis=1)
df.head(10)
```

```python
bc = []
bs = []
art_iurl = []
art_ixurl = []
art_l = []
art_pc = []
art_t = []
for i, r in df_ar.iterrows():
    bio = literal_eval(r['bio'])
    image = literal_eval(r['image'])
    stats = literal_eval(r['stats'])
    tags = literal_eval(r['tags'])

    ail = 'n/a'
    aiexl = 'n/a'
    at = []
    if 'summary' in bio:
        bs.append(bio['summary'])
    else:
        bs.append('n/a')
    if 'content' in bio:
        bc.append(bio['content'])
    else:
        bc.append('n/a')
    for k in image:
        if k['size'] == 'large':
            ail = k['#text']
        if k['size'] == 'extralarge':
            aiexl = k['#text']
    for art in tags:
        at.append(art['name'])
    
    if 'playcount' in stats:
        art_pc.append(stats['playcount'])
    else:
        art_pc.append(0)
    if 'listeners' in stats:
        art_l.append(stats['listeners'])
    else:
        art_l.append(0)
    art_iurl.append(ail)
    art_ixurl.append(aiexl)
    art_t.append(at)
print(len(art_pc))
```

```python
df_artist = pd.DataFrame({'artist_name': df_artist_name, 'content': bc, 'summary': bs, 'artist_listeners': art_l, 'artist_playcount': art_pc, 'artist_tag': art_t, 'artist_image_url_large': art_iurl, 'artist_image_url_exlarge': art_ixurl})
df_artist.head(5)
```
