## Observations

1. errors observed with many artist names and requires intense cleaning to get artist information accurately.
2. Also it will be difficult to append to the panda data frame as artists are unique after converting it into set

```python
r2=[]
r3=[]
r4=[]
r5=[]
for i in artist_list_df[0:10000] :

    r = lastfm_get({
        'method': 'artist.getInfo',
        'artist': i
    })
    #r.status_code
    r_json = r.json()
    if 'error' in r_json:
        r2.append({})
        r3.append([])
        r4.append({})
        r5.append([])
    else:
        if 'bio' in r_json['artist']:
            
            r_bio = r_json['artist']['bio']
        else:
            r_bio = {}
        if 'image' in r_json['artist']:
            r_image=r_json['artist']['image']
        else:
            r_image = []
        if 'stats' in r_json['artist']:
            r_stats=r_json['artist']['stats']
        else:
            r_stats={}
        if 'tags' in r_json['artist']:
            r_tags=r_json['artist']['tags']['tag']
        else:
            r_tags=[]
        
        r2.append(r_bio)
        r3.append(r_image)
        r4.append(r_stats)
        r5.append(r_tags)

#r_df=pd.DataFrame(r5)
r_df = pd.DataFrame({'bio':r2,'image':r3,'stats':r4,'tags':r5})
#print(r_df)
export= r_df.to_csv('artists.csv')

print(len(r2),len(r3),len(r4),len(r5))
```
